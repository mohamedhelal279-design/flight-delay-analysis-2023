```python
flights = pd.read_csv(r"E:\Ahmed fathy\introduction to Data analytics\introduction to Data analytics\US 2023 civil flights\US_flights_2023.csv")
weather = pd.read_csv(r"E:\Ahmed fathy\introduction to Data analytics\introduction to Data analytics\US 2023 civil flights\weather_meteo_by_airport.csv")
cancelled_diverted = pd.read_csv(r"E:\Ahmed fathy\introduction to Data analytics\introduction to Data analytics\US 2023 civil flights\Cancelled_Diverted_2023.csv")
airports_location = pd.read_csv(r"E:\Ahmed fathy\introduction to Data analytics\introduction to Data analytics\US 2023 civil flights\airports_geolocation.csv")
```

```python
day_of_week_mapping = {
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday",
    4: "Thursday",
    5: "Friday",
    6: "Saturday",
    7: "Sunday",
}
flights["Day_Of_Week"] = flights["Day_Of_Week"].map(day_of_week_mapping)
cancelled_diverted["Day_Of_Week"] = cancelled_diverted["Day_Of_Week"].map(day_of_week_mapping)
```

```python
flights["FlightDate"] = flights["FlightDate"].astype("datetime64[ns]")
```

```python
categorical_columns =[
    "Airline",
    "Dep_Airport",
    "Dep_CityName",
    "DepTime_label",
    "Dep_Delay_Type",
    "Arr_Airport",
    "Arr_CityName",
    "Arr_Delay_Type",
    "Distance_type",
    "Manufacturer",
    "Model",
    "Day_Of_Week"
]
for column in categorical_columns:
    flights[column] = flights[column].astype("category")
```

```python
num_columns = [
    "Dep_Delay",
    "Dep_Delay_Tag",
    "Arr_Delay",
    "Flight_Duration",
    "Delay_Carrier",
    "Delay_Weather",
    "Delay_NAS",
    "Delay_Security",
    "Delay_LastAircraft",
    "Aicraft_age",
]
for column1 in num_columns :
    flights[column1]=flights[column1].astype("int32")
```

```python
def normalize_columns(df):
    df.columns =df.columns.str.lower().str.replace(" ","_")
    return df
```

```python
flights = normalize_columns(flights)
cancelled_diverted= normalize_columns(cancelled_diverted)
weather = normalize_columns(weather)
airports_location = normalize_columns(airports_location)
```
